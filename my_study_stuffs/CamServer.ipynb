```python
import socket
import struct
import pandas as pd
import pickle
import cv2
import numpy as np

HOST=''
PORT=8089

cam = cv2.VideoCapture(0)
width=cam.get(cv2.CAP_PROP_FRAME_WIDTH)
height=cam.get(cv2.CAP_PROP_FRAME_HEIGHT)
print('orig size : %d, %d' % (width, height))

cam.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cam.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
print('Socket created')

server.bind((HOST, PORT))
print('Socket bind complete')

server.listen(10)
print('Socket now listening')

server_cam, addr = server.accept()
print('New Client Comming..')

## 카메라 찍어서 전달한다.
while True:
    cmd_byte = server_cam.recv(1)
    cmd = struct.unpack('!B', cmd_byte)
    print(cmd[0])
    if cmd[0] == 12:
        print('12 OK .. read and send ...')
        ret, frame = cam.read()
        status, frame = cv2.imencode('.jpg', frame, [cv2.IMWRITE_JPEG_QUALITY, 30])
        print('cam read ...')
        data = pickle.dumps(frame)
        print('data len : ' + str(len(data)))
        data_size = struct.pack('!L', len(data))
        server_cam.sendall(data_size + data)
                            
server_cam.close()
server.close()
```
